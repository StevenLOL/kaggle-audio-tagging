```python
%%time

TRAIN_CSV = '../../data/external/train.csv'
TRAIN_FILES = '../../data/external/audio_train'

train_files = find_paths_with_tags(csv_path=TRAIN_CSV,
                                   files_path=TRAIN_FILES, limit=None)
```

```python
%%time
train_wav_inputs = load_wav_files(train_files, duration=15)
```

```python
%%time
f'{sys.getsizeof(train_wav_inputs)} bytes'
```

```python
def chunkIt(og:list, chunks:int=1):
  return [og[i::chunks] for i in range(chunks)]

split_inputs = chunkIt(train_wav_inputs, int(len(train_wav_inputs)/1000))
list(map(lambda x: f'{sys.getsizeof(x)} bytes', split_inputs))
```

```python
%%time

for i, inputs in enumerate(split_inputs):
    _tmp_wavs = [ x.__dict__ for x in inputs ]
    serialized_inputs = serialize_wavs(_tmp_wavs)
    pickle_out(f'train_wavs_{i}.pkl', serialized_inputs)
```

```python
%%time

with open('train_wavs_1.pkl', 'rb') as f:
    data = pickle.loads(f.read())

print(f'{sys.getsizeof(data)} bytes')
deserialized_train_wav_inputs = deserialize_wavs(data)
print(f'{sys.getsizeof(deserialized_train_wav_inputs)} bytes')
```
